```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

# Creating data
s = np.linspace(-10, 10, 10)
t = np.linspace(-10, 10, 10)
S, T = np.meshgrid(s, t)
Z = np.zeros(S.shape)

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(S, T, Z, alpha=0.5, color='blue')
ax.set_xlabel('S-axis')
ax.set_ylabel('T-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Space spanned by [s, t, 0]')
plt.show()
```
